```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'AREA', data = Crime_Data_from_2020_to_Present, color = 'blue', 
              order = Crime_Data_from_2020_to_Present['AREA'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'AREA NAME', data = Crime_Data_from_2020_to_Present, color = 'blue', 
              order = Crime_Data_from_2020_to_Present['AREA NAME'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'Rpt Dist No', data = Crime_Data_from_2020_to_Present, color = 'blue', 
              order = Crime_Data_from_2020_to_Present['Rpt Dist No'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'Crm Cd', data = Crime_Data_from_2020_to_Present, color = 'blue', 
              order = Crime_Data_from_2020_to_Present['Crm Cd'].value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'Crm Cd Desc', data = Crime_Data_from_2020_to_Present, color = 'blue', 
              order = Crime_Data_from_2020_to_Present['Crm Cd Desc'].head(20).value_counts().index);
sns.countplot(ax = axes[2, 1], x = 'Mocodes', data = Crime_Data_from_2020_to_Present, color = 'blue', 
              order = Crime_Data_from_2020_to_Present['Mocodes'].head(20).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```

```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

```python
log_transform(Crime_Data_from_2020_to_Present,['Premis Cd','Vict Age'])
```
